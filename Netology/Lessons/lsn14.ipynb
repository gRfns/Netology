Описание столбцов
surgery - 
была операция 
1 = Да 
2 = Выздровела без операции

hospital_number - Номер "дела" присвоеный лошади (не может быть уникальным если лошадь лечилась больше однго раза

age - возвраст лошади 1 = Взрослая лошадь 2 = молодая ( 6 < месяцев)

rectal_temp - ректальная температура, норма 37.8,
* Повышенная температура может возникнуть из-за инфекции.
* температура может быть снижена, когда животное находится в позднем шоке
* этот параметр обычно меняется по мере развития проблемы
например, может начаться нормально, а затем стать повышенным из-
за поражения, проходя обратно через нормальный диапазон, когда
лошадь впадает в шок

pulse - частота сердечных сокращений в ударах в минуту, норма 30-40
* очень редко может быть низкий пульс(исключения спортивные лошади)
* животные с болезненными поражениями или страдающие от кровеносного шока
могут иметь повышенную частоту сердечных сокращений

abdominal_distension - вздутие живота
 ВАЖНЫЙ параметр.
          - возможные значения
               1 = нет
               2 = легкий
               3 = умеренный
               4 = тяжелая
          - животное с вздутием живота может причинить боль и
            имеют сниженную перистальтику кишечника.
          - лошади с сильным вздутием живота, вероятно, потребуются
            операция просто сними давление

respiratory_rate - частота дыхания, норма 8-10,
* полезность сомнительна из-за больших колебаний

pain - боль, чем больнее тем больше вероятность что нужна хирургия
1 = бдительность, отсутствие боли
2 = депрессия
3 = прерывистая легкая боль
4 = прерывистая сильная боль
5 = непрерывная сильная боль

outcome - Что случилось с лошадью
1 = жил
2 = умер
3 = был усыплен

#### Работаем с выбросами

```python
my8.info()
```

```python
q1 = my8['rectal_temp'].quantile(0.25)
q3 = my8['rectal_temp'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = my8[my8['rectal_temp'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([my8, remove_outliers]).drop_duplicates(keep=False).dropna()
```

```python
q1 = my8['respiratory_rate'].quantile(0.25)
q3 = my8['respiratory_rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = my8[my8['respiratory_rate'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([my8, remove_outliers]).drop_duplicates(keep=False).dropna()
```

```python
q1 = my8['pulse'].quantile(0.25)
q3 = my8['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = my8[my8['pulse'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([my8, remove_outliers]).drop_duplicates(keep=False).dropna()
```
